# Import Libraries

```python
import os
import streamlit as st
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import Chroma
from langchain.chains import RetrievalQA
from langchain_openai import OpenAI
from langchain_openai import OpenAIEmbeddings

from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
```

## Intializing OpenAI LLM

```python
os.environ.get('OPENAI_API_KEY')
```

```python
llm = OpenAI(openai_api_key=os.environ.get('OPENAI_API_KEY'))
```

## Intializing OpenAI Embeddings

```python
embeddings = OpenAIEmbeddings(openai_api_key=os.environ.get('OPENAI_API_KEY'))
```

## Loading Quran (With Text Loader)

```python
help(CharacterTextSplitter)
```

```python
loader = TextLoader("quran.txt",encoding='utf8')
text_splitter = CharacterTextSplitter(chunk_size=256, chunk_overlap=128,separator='')
documents = loader.load()
```

```python
documents
```

```python
docs = text_splitter.split_documents(documents)
```

```python
pages = docs
```

```python
len(docs[0].page_content)
```

## Loading Quran (With PDF Loader)

```python
loader = PyPDFLoader("quran.pdf")
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1024, chunk_overlap=512)

pages = loader.load_and_split(text_splitter=text_splitter)
```

```python
len(pages)
```

```python
pages[5]
```

## Intializaing Chroma Vector Store

```python
vector_store = Chroma.from_documents(pages, embeddings, persist_directory="./chroma_db")
```

```python
vector_store = Chroma(persist_directory="./chroma_db", embedding_function=embeddings)
```

```python
retriever=vector_store.as_retriever()
chain = RetrievalQA.from_chain_type(llm, retriever=retriever)
```

```python
question = 'who is austin'
response = chain.run(question)
print(response)
```
